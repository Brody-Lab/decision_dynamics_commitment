```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
import matplotlib.colors as colors
import matplotlib
```

# Figure 1d

```python
fig, ax = plt.subplots(figsize=(3, 3))

resolution = 11
x_, y_ = np.meshgrid(
    np.linspace(-1, 1, num=resolution), 
    np.linspace(-1, 1, num=resolution)
)

u_ = np.zeros_like(x_)
v_ = np.zeros_like(y_)
u_left_ = np.zeros_like(x_)
v_left_ = np.zeros_like(y_)
u_right_ = np.zeros_like(x_)
v_right_ = np.zeros_like(y_)

for i in range(resolution):
    for j in range(resolution):
        u_[i, j] = 10 * np.sin(np.pi*x_[i, j]/0.7)
        v_[i, j] = -30*y_[i, j]
        u_left_[i, j] = -1
        v_left_[i, j] = 0
        u_right_[i, j] = 1
        v_right_[i, j] = 0

ax.set_aspect('equal')
ax.quiver(x_, y_, u_, v_, angles='xy', scale_units='xy', scale=100)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(1)
ax.spines['left'].set_linewidth(1)
ax.set_xticks([])
ax.set_yticks([])
ax.tick_params(axis='both', which='major', labelsize=25)
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
```

```python
fig, ax = plt.subplots(figsize=(3, 3))

resolution = 100
x, y = np.meshgrid(
    np.linspace(-1, 1, num=resolution), 
    np.linspace(-1, 1, num=resolution)
)

u = np.zeros_like(x)
v = np.zeros_like(y)
u_left = np.zeros_like(x)
v_left = np.zeros_like(y)
u_right = np.zeros_like(x)
v_right = np.zeros_like(y)
energy = np.zeros_like(x)
energy_input = np.zeros_like(x)

for i in range(resolution):
    for j in range(resolution):
        u[i, j] = 2 * x[i, j] * (0.49 - x[i, j]**2)
        v[i, j] = -2 * y[i, j]
        energy[i, j] = np.sqrt(u[i, j]**2 + v[i, j]**2)
        u_left[i, j] = -1
        v_left[i, j] = 0
        u_right[i, j] = 1
        v_right[i, j] = 0
        energy_input[i, j] = (
            np.sqrt(u_left[i, j]**2 + v_left[i, j]**2) + \
            np.sqrt(u_right[i, j]**2 + v_right[i, j]**2)
        )/2        
        
ax.set_aspect('equal')
c1 = ax.pcolormesh(
    x,
    y,
    energy, 
    cmap='gray', 
    norm=matplotlib.colors.LogNorm(vmin=0.01, vmax=1)
)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(1)
ax.spines['left'].set_linewidth(1)
ax.set_xticks([])
ax.set_yticks([])
ax.tick_params(axis='both', which='major', labelsize=25)
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])

cbar = fig.colorbar(c1, ax=ax, fraction=0.045, pad=0.04)
cbar.ax.set_yscale('log')
cbar.ax.set_yticks([])
cbar.ax.tick_params(labelsize=25)
```

# Figure 1g

```python
fig, ax = plt.subplots(figsize=(3, 3))

resolution = 11
x_, y_ = np.meshgrid(
    np.linspace(-1, 1, num=resolution), 
    np.linspace(-1, 1, num=resolution)
)

u_ = np.zeros_like(x_)
v_ = np.zeros_like(y_)
u_left_ = np.zeros_like(x_)
v_left_ = np.zeros_like(y_)
u_right_ = np.zeros_like(x_)
v_right_ = np.zeros_like(y_)

for i in range(resolution):
    for j in range(resolution):
        if x_[i, j] >= 0.7 or x_[i, j] <= -0.7:
            u_[i, j] = 1/3 * x_[i, j] * (0.49 - x_[i, j]**2)
            v_[i, j] = -1*y_[i, j]
        else:
            u_[i, j] = 0
            v_[i, j] = -1*y_[i, j]
        u_left_[i, j] = -1
        v_left_[i, j] = 0
        u_right_[i, j] = 1
        v_right_[i, j] = 0

ax.set_aspect('equal')
ax.quiver(x_, y_, u_, v_, angles='xy', scale_units='xy', scale=3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(1)
ax.spines['left'].set_linewidth(1)
ax.set_xticks([])
ax.set_yticks([])
ax.tick_params(axis='both', which='major', labelsize=25)
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
```

# Figure 1h

```python
fig, ax = plt.subplots(figsize=(3, 3))

resolution = 11
x_, y_ = np.meshgrid(
    np.linspace(-1, 1, num=resolution), 
    np.linspace(-1, 1, num=resolution)
)

u_ = np.zeros_like(x_)
v_ = np.zeros_like(y_)
u_left_ = np.zeros_like(x_)
v_left_ = np.zeros_like(y_)
u_right_ = np.zeros_like(x_)
v_right_ = np.zeros_like(y_)

for i in range(resolution):
    for j in range(resolution):
        u_[i, j] = y_[i, j]
        v_[i, j] = -y_[i, j]
        u_left_[i, j] = 0
        v_left_[i, j] = -1
        u_right_[i, j] = 0
        v_right_[i, j] = 1

ax.set_aspect('equal')
ax.quiver(x_, y_, u_, v_, angles='xy', scale_units='xy', scale=5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(1)
ax.spines['left'].set_linewidth(1)
ax.set_xticks([])
ax.set_yticks([])
ax.tick_params(axis='both', which='major', labelsize=25)
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
```
