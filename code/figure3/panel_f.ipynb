# Load libraries

```python
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
import matplotlib.colors as colors
from sklearn.model_selection import KFold
import numpy as np

from sklearn.decomposition import PCA
from scipy.ndimage.filters import gaussian_filter
import csv
```

```python
data = np.load('../../data/processed_data/extended_data_figure_5/findr_vs_cfindr_all_sessions.npz')
second_quartile_r2_findr_priors=data['second_quartile_r2_findr_priors']
second_quartile_r2_findr_priors_c=data['second_quartile_r2_findr_priors_c']
r2_findr_priors_sess=data['r2_findr_priors_sess']
r2_findr_priors_c_sess=data['r2_findr_priors_c_sess']
```

```python
fig, ax = plt.subplots(figsize=(6, 5), constrained_layout=True)
ax.hist(r2_findr_priors_sess - r2_findr_priors_c_sess, bins=30, color='lightgray', edgecolor='k')
ax.set_xlim([-0.5, 1])
ax.vlines(x=0, ymin=0, ymax=700, color='k', linestyle='--')

ax.spines[['right', 'top']].set_visible(False)
ax.set_xlabel("FINDR − constrained prior model \n(median evidence sign-conditioned PSTH $R^2$)", fontsize=20)
ax.set_ylabel("neurons", fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)
```

```python
fig, ax = plt.subplots(figsize=(6, 5), constrained_layout=True)
ax.hist(
    np.array(second_quartile_r2_findr_priors) - np.array(second_quartile_r2_findr_priors_c), 
    bins=9, 
    color='lightgray', 
    edgecolor='k'
)
ax.vlines(x=0, ymin=0, ymax=8, color='k', linestyle='--')

ax.spines[['right', 'top']].set_visible(False)
ax.set_xlabel("FINDR − constrained prior model \n(median $\Delta R^2$)", fontsize=20)
ax.set_ylabel("sessions", fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)
```
